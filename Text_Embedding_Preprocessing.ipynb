```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
print("Tokenizer loaded successfully!")
```

```python
texts = [
    "A red rose with green leaves.",
    "A yellow sunflower under bright sunlight.",
    "A white daisy in a garden."
]

print("Sample texts loaded successfully!")
print(texts)
```

```python
encoded = tokenizer(
    texts,
    padding=True,
    truncation=True,
    return_tensors="pt"
)

print("Input IDs shape:", encoded["input_ids"].shape)
print("Attention Mask shape:", encoded["attention_mask"].shape)
```

```python
print("Input IDs:")
print(encoded["input_ids"])

print("\nAttention Mask:")
print(encoded["attention_mask"])
```

```python
decoded = tokenizer.decode(encoded["input_ids"][0])
print("Decoded text:")
print(decoded)
```

```python
for i, text in enumerate(texts):
    tokens = tokenizer.tokenize(text)
    print(f"\nText {i+1}: {text}")
    print("Tokens:", tokens)
    print("Token count:", len(tokens))
```
