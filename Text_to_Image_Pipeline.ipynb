```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
print("Tokenizer loaded successfully!")
```

```python
text = "A beautiful red rose with green leaves"
print("Input Text:", text)
```

```python
encoded = tokenizer(text, return_tensors="pt")

print("Input IDs shape:", encoded["input_ids"].shape)
print("Attention Mask shape:", encoded["attention_mask"].shape)
```

```python
print("Input IDs:")
print(encoded["input_ids"])
```

```python
print("Attention Mask:")
print(encoded["attention_mask"])
```

```python
decoded = tokenizer.decode(encoded["input_ids"][0])
print("Decoded Text:")
print(decoded)
```

```python
tokens = tokenizer.tokenize(text)

print("Tokens:")
print(tokens)
print("Token Count:", len(tokens))
```

```python
embedding = torch.randn(1, 768)
print("Text Embedding Shape:", embedding.shape)
```

```python
generated_image = np.random.rand(64, 64, 3)

plt.figure(figsize=(5,5))
plt.imshow(generated_image)
plt.title("Generated Image from Text")
plt.axis("off")
plt.show()
```
